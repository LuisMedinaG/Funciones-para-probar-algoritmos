```
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```
def plotFunc(min_x, max_x, min_y, max_y, s, f):
    X = np.arange(min_x, max_x, s)
    Y = np.arange(min_y, max_y, s)
    X, Y = np.meshgrid(X, Y)
    Z = f(X, Y)
    
    fig = plt.figure()
    ax = fig.gca(projection = '3d')
    
    surf = ax.plot_surface(X, Y, Z, cmap = cm.coolwarm, linewidth = 1)

    fig.colorbar(surf)
    plt.show()
```

```
def Sphere(X, Y):
    return X**2 + Y**2 

def Rosenbrock(X, Y):
    return 100 * (Y - X ** 2) ** 2 + (1 - X) ** 2

def Rastrigin(X, Y):
    return (X**2 - 10 * np.cos(2 * np.pi * X)) + (Y**2 - 10 * np.cos(2 * np.pi * Y))

def Quartic(X, Y):
    return X**4 + Y**4
```

```
plotFunc(-5.12, 5.12, -5.12, 5.12, 0.1, Sphere)
```

```
plotFunc(-2.048, 2.048, -1, 3, 0.1, Rosenbrock)
```

```
plotFunc(-5.12, 5.12, -5.12, 5.12, 0.1, Rastrigin)
```

```
plotFunc(-1.28, 1.28, -1.28, 1.28, 0.1, Quartic)
```
